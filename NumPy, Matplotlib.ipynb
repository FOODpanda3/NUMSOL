```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x): return x**5 - x - 1
def f2(x): return x**5 - 5*x**3 + 4*x + 1
def f3(x): return x**5 + 2*x**4 - x - 3
def f4(x): return x**5 - 10
def f5(x): return x**5 - 2*x**2 + x - 0.5

x = np.linspace(-3, 3, 1000)

functions = [f1, f2, f3, f4, f5]
titles = [
    r"$f_1(x) = x^5 - x - 1$",
    r"$f_2(x) = x^5 - 5x^3 + 4x + 1$",
    r"$f_3(x) = x^5 + 2x^4 - x - 3$",
    r"$f_4(x) = x^5 - 10$",
    r"$f_5(x) = x^5 - 2x^2 + x - 0.5$"
]

plt.figure(figsize=(12, 10))

for i, f in enumerate(functions):
    plt.subplot(3, 2, i+1)
    plt.plot(x, f(x))
    plt.axhline(0)
    plt.grid(True)
    plt.title(titles[i])

plt.tight_layout()
plt.show()

# ==========================================
# PART B: Bisection Method Logic
# ==========================================
def bisection(f, a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("Invalid interval: f(a) and f(b) must have opposite signs.")

    iterations = []
    
    for i in range(max_iter):
        c = (a + b) / 2
        iterations.append((i+1, a, b, c, f(c)))
        
        if abs(f(c)) < tol or (b - a)/2 < tol:
            break
        
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
            
    return c, iterations
    
root_f1, data_f1 = bisection(f1, 1, 2)

print(f"Approximate root: {root_f1:.6f}")
print("Iteration | a | b | c | f(c)")
for row in data_f1[:5]:
    print(row)

root_f2, data_f2 = bisection(f2, -3, -2)

print(f"Approximate root: {root_f2:.6f}")
print("Iteration | a | b | c | f(c)")
for row in data_f2[:5]:
    print(row)

root_f3, data_f3 = bisection(f3, 1, 2)

print(f"Approximate root: {root_f3:.6f}")
print("Iteration | a | b | c | f(c)")
for row in data_f3[:5]:
    print(row)

root_f4, data_f4 = bisection(f4, 1, 2)

print(f"Approximate root: {root_f4:.6f}")
print("Iteration | a | b | c | f(c)")
for row in data_f4[:5]:
    print(row)

root_f5, data_f5 = bisection(f5, 1, 2)

print(f"Approximate root: {root_f5:.6f}")
print("Iteration | a | b | c | f(c)")
for row in data_f5[:5]:
    print(row)
# ==========================================
# PART C: Execution and Convergence Tracking
# ==========================================
iterations = [row[0] for row in data_f1]
midpoints = [row[3] for row in data_f1]

plt.plot(iterations, midpoints, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Midpoint Approximation")
plt.title("Bisection Convergence for f1(x)")
plt.grid(True)
plt.show()

errors = [abs(row[4]) for row in data_f1]

plt.semilogy(iterations, errors, marker='o')
plt.xlabel("Iteration")
plt.ylabel("|f(c)|")
plt.title("Error Convergence (Log Scale)")
plt.grid(True)
plt.show()

iterations = [row[0] for row in data_f2]
midpoints = [row[3] for row in data_f2]

plt.plot(iterations, midpoints, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Midpoint Approximation")
plt.title("Bisection Convergence for f2(x)")
plt.grid(True)
plt.show()

errors = [abs(row[4]) for row in data_f2]

plt.semilogy(iterations, errors, marker='o')
plt.xlabel("Iteration")
plt.ylabel("|f(c)|")
plt.title("Error Convergence (Log Scale)")
plt.grid(True)
plt.show()

iterations = [row[0] for row in data_f3]
midpoints = [row[3] for row in data_f3]

plt.plot(iterations, midpoints, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Midpoint Approximation")
plt.title("Bisection Convergence for f3(x)")
plt.grid(True)
plt.show()

errors = [abs(row[4]) for row in data_f3]

plt.semilogy(iterations, errors, marker='o')
plt.xlabel("Iteration")
plt.ylabel("|f(c)|")
plt.title("Error Convergence (Log Scale)")
plt.grid(True)
plt.show()

# ==========================================
# PART D: Convergence Visualization
# ==========================================
def plot_convergence(eq_name):
    data = all_results[eq_name]
    iters = [row[0] for row in data]
    midpoints = [row[3] for row in data]
    errors = [abs(row[4]) for row in data]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

    # Midpoint Convergence (Task C1)
    ax1.plot(iters, midpoints, 'b-o', markersize=4)
    ax1.set_title(f"Midpoint Path: {eq_name}")
    ax1.set_xlabel("Iteration")
    ax1.set_ylabel("Root Approximation (c)")
    ax1.grid(True)

    # Error Convergence (Task C2)
    ax2.semilogy(iters, errors, 'r-s', markersize=4)
    ax2.set_title(f"Error Decay (Log Scale): {eq_name}")
    ax2.set_xlabel("Iteration")
    ax2.set_ylabel("|f(c)|")
    ax2.grid(True)
    plt.show()

# Visualize for at least two equations as per Student Task
plot_convergence("f1")
plot_convergence("f2")
```
