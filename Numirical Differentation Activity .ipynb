```python
import numpy as np

def f1(x):
    return (x**2 + 1) / np.sin(x)

x = 7
dx = 0.01

# Calculations
f_x = f1(x)
f_plus = f1(x + dx)
f_minus = f1(x - dx)

fdd = (f_plus - f_x) / dx
bdd = (f_x - f_minus) / dx
cdd = (f_plus - f_minus) / (2 * dx)

print(f"FDD: {fdd:.5f}")
print(f"BDD: {bdd:.5f}")
print(f"CDD: {cdd:.5f}")
```

```python
import numpy as np

def f2(x):
    return np.sin(x) / (x**3 - 3 * x**5)

x = 7
dx = 0.01

# Calculations
f_x = f2(x)
f_plus = f2(x + dx)
f_minus = f2(x - dx)

fdd = (f_plus - f_x) / dx
bdd = (f_x - f_minus) / dx
cdd = (f_plus - f_minus) / (2 * dx)

print(f"FDD: {fdd:.10f}")
print(f"BDD: {bdd:.10f}")
print(f"CDD: {cdd:.10f}")
```

```python
import numpy as np

def f3(x):
    numerator = np.cos(x) * np.sin(x)
    denominator = 3 * x**2 + (np.cos(x + 2) / x**2)
    return numerator / denominator

x = 7
dx = 0.01

# Calculations
f_x = f3(x)
f_plus = f3(x + dx)
f_minus = f3(x - dx)

fdd = (f_plus - f_x) / dx
bdd = (f_x - f_minus) / dx
cdd = (f_plus - f_minus) / (2 * dx)

print(f"FDD: {fdd:.10f}")
print(f"BDD: {bdd:.10f}")
print(f"CDD: {cdd:.10f}")
```
