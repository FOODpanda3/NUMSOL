```python
import numpy as np

def custom_gaussian_elimination(A, b):
    # Convert to float to ensure precision during division
    A = A.astype(float)
    b = b.astype(float).reshape(-1, 1)
    n = len(b)
    
    # Combine into an augmented matrix
    augmented = np.hstack((A, b))
    
    print("Step 0: Initial Augmented Matrix")
    print(augmented)
    print("-" * 30)

    # --- Forward Elimination ---
    for i in range(n):
        # Pivoting to ensure the largest absolute value is at the pivot position
        pivot_row = np.argmax(np.abs(augmented[i:, i])) + i
        augmented[[i, pivot_row]] = augmented[[pivot_row, i]]
        
        for j in range(i + 1, n):
            factor = augmented[j, i] / augmented[i, i]
            # Zero out the element below the pivot
            augmented[j, i:] = augmented[j, i:] - factor * augmented[i, i:]
            
            print(f"Eliminating column {i}, row {j}:")
            print(np.round(augmented, 4))
            print("-" * 20)

    # --- Back Substitution ---
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = (augmented[i, -1] - np.dot(augmented[i, i+1:n], x[i+1:n])) / augmented[i, i]
        
    return x

# Problem Setup
A1 = np.array([[2, 1, 1], [-1, 2, -1], [3, -1, 2]])
b1 = np.array([36, -11, 47])

# Run Solver
m_packets = custom_gaussian_elimination(A1, b1)

print("Decrypted Data Packets (m1, m2, m3):")
print(m_packets)
```
